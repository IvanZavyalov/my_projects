- В годе релиза данные представлены типом float64,что  связано с наличием пропусков в данных.Заменим пропуски на значение-индикатор 0, чтобы не потерять строки с данными, так как число пропусков более 1,5 %. Года релиза со значением-индикатором не войдут в итоговый срез данных.  Приведём столбец к целочисленному формату,для удобства последующей фильтрации.

```python
df['year_of_release']=df['year_of_release'].fillna(0)
```

- Выведем на экран обновлённые значения года релиза.

- Удалим строки с пропусками в столбцах с названием игр и жанром.

```python
df=df.dropna(subset=['name','genre'])
```

- Выведем обновлённую информацию о датасете.

```python
df.info()
```

- Заменим пропуски в столбцах с продажами в Европе и Японии средним значением, в зависимости от названия платформы и года выхода игры. Для этого применим метод apply().

```python
def mean_group_eu_sales(row):
    """Функция принимает аргумент row-строка из датафрейма df,
    создаёт переменную group в которую сохраняет отфильтрованный датафрейм по 2 условиям:
    -год релиза совпадает с годом релиза в текущей строке
    -название платформы совпадает с названием платформы в текущей строке.
    В конце функция возвращает среднее значение продаж """
    
    group = df[(df['year_of_release'] == row['year_of_release']) & 
               (df['platform'] == row['platform'])]
    return group['eu_sales'].mean()
df['eu_sales']=df.apply(mean_group_eu_sales,axis=1)
```

```python
def mean_group_jp_sales(row):
    """Функция принимает аргумент row-строка из датафрейма df,
    создаёт переменную group в которую сохраняет отфильтрованный датафрейм по 2 условиям:
    -год релиза совпадает с годом релиза в текущей строке
    -название платформы совпадает с названием платформы в текущей строке.
    В конце функция возвращает среднее значение продаж """
    
    group = df[(df['year_of_release'] == row['year_of_release']) & 
               (df['platform'] == row['platform'])]
    return group['jp_sales'].mean()
df['jp_sales']=df.apply(mean_group_jp_sales,axis=1)
```

В столбцах с названиеями игр и жанрами было 2 общих пропуска и строки с ними были удалены. Пропуски в столбце с годом релиза были заменены на значение-индикатор 0, который не сказался на дальнейшем анализе. Пропуски в столбцах с количеством продаж в Европе и Японии были заменены среднем значением с учётом платформы и года релиза. Пропуски в столбцах с оценками критиков и пользователей, а также рейтингом остались без изменений. После очистки данных для исследования был сделан срез за период с 2000 по 2013 года включительно. Данные были категоризированы по оценкам критиков и пользователей на 3 группы:'Низкая оценка','Средняя оценка' и 'Высокая оценка'. В результате было выявлено, что наибольшее количество игр в этом периоде имеют среднюю и высокую оценки, что указывает на высокий уровень выпущенных игр. Также были выделены топ-7 платформ по количеству выпущенных игр, самые популярные из них: 'PS2','DS' и 'Wii'.
